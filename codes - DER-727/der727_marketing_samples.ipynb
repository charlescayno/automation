```python
def export_raw_df_to_another_excel(unit, input_folder, output_foler, input_filename, output_filename, input_sheet_name, output_sheet_name, anchor="B2"):

    input_path = path_maker(f'{input_folder}')
    output_path = input_path + output_filename    
    input_file = input_folder + input_filename

    df = pd.read_excel(input_file, input_sheet_name)

    if input_sheet_name == "Resistor Dimming":
        df = clean_dataframe(df)

        df_header_list = ['Resistor', 'led_load', 'Vin', 'Freq (Hz)', 'Vac (VAC)', 'Iin (mA)', 'Pin (W)', 'PF', 'THD (%)', 'Vo (V)', 'Io1 (mA)', 'Po (W)', 'Vreg (%)', 'Ireg (%)', 'Eff (%)',
                            'Resistor', 'led_load', 'Vin', 'Freq (Hz)', 'Vac (VAC)', 'Iin (mA)', 'Pin (W)', 'PF', 'THD (%)', 'Vo (V)', 'Io1 (mA)', 'Po (W)', 'Vreg (%)', 'Ireg (%)', 'Eff (%)',
                            'Resistor', 'led_load', 'Vin', 'Freq (Hz)', 'Vac (VAC)', 'Iin (mA)', 'Pin (W)', 'PF', 'THD (%)', 'Vo (V)', 'Io1 (mA)', 'Po (W)', 'Vreg (%)', 'Ireg (%)', 'Eff (%)',
                            'Resistor', 'led_load', 'Vin', 'Freq (Hz)', 'Vac (VAC)', 'Iin (mA)', 'Pin (W)', 'PF', 'THD (%)', 'Vo (V)', 'Io1 (mA)', 'Po (W)', 'Vreg (%)', 'Ireg (%)', 'Eff (%)']
        df_header = create_header_list(df_header_list)
        export_df_to_excel(df_header, input_folder, output_filename, output_sheet_name, anchor="B1")
        export_df_to_excel(df, input_folder, output_filename, output_sheet_name, anchor=anchor)
    
    
    
    elif output_sheet_name == "Harmonics":
        df = select_dataframe(df, start_anchor="E9", end_anchor="P37")
        export_df_to_excel(df, input_folder, output_filename, output_sheet_name, anchor=anchor)

    elif output_sheet_name == "Parametrics":
        export_df_to_excel(df, input_folder, output_filename, output_sheet_name, anchor=anchor)
    
    else: # default case
        df = clean_dataframe(df)
        export_df_to_excel(df, input_folder, output_filename, output_sheet_name, anchor=anchor)
    
    return df
```

```python
def df_by_index_led(df_dict, index, led):
    df = df_dict[index]
    df = df.loc[df['led_load'] == led]
    return df
```
